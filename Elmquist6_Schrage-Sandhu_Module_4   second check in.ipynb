## Disease Background:

Disease Background: Mpox in the Dominican Republic
Prevalence & Incidence

- In the Dominican Republic (DR), from July 2022 to February 2023, 71 confirmed mpox cases were reported out of 283 suspected cases. 
- During that same outbreak, a surprisingly high proportion of cases were women (32.4%) and children under 10 years old (22.5%), which is distinct from global trends. 
- Cases in the DR peaked around September 2022 and declined thereafter. 
- According to the Pan American Health Organization (PAHO), by epidemiological week 40 of 2024, the DR accounted for 66% of mpox cases in the Caribbean and Atlantic islands subregion. 
- As of August 2024, the DR reported no new mpox cases since April, per the Ministry of Health.

Economic Burden
- Globally, mpox treatment is primarily supportive care, so costs include symptom management (pain, rash), hydration, skin care, prevention and treatment of secondary bacterial infections, plus possible hospitalization for severe cases. 
- In resource-limited settings (like parts of the Caribbean), limited diagnostic capacity (e.g., limited access to PCR testing) can increase costs and delay detection, which burdens public health systems. 
- Indirect economic costs may come from isolation, loss of productivity (work/school), and public health surveillance efforts, particularly during outbreaks.
- There’s also potential cost associated with vaccination programs (pre- or post-exposure), though specific DR vaccine-deployment costs aren’t well documented in the literature.
- Stigma and fear may indirectly contribute to cost — if people delay seeking care, outbreaks can spread and require more resources to manage.


Risk Factors & Societal Determinants
- According to meta-analysis, HIV infection greatly increases risk (OR ≈ 4.46). 
- Other sexually transmitted infections (STIs), unprotected sexual activity, and having multiple sexual partners are also significant risk factors. 
- Identifying as men who have sex with men (MSM) is associated with higher risk in many settings (OR ≈ 2.18 in some studies). 
- Contact with infected individuals (through close skin-to-skin or household contact) strongly contributes to transmission. 
- Interaction with infected animals (especially in zoonotic settings) is a risk factor in endemic regions (OR ≈ 5.6). 
- Lack of smallpox vaccination: prior smallpox vaccination is protective (meta-analysis shows reduced risk). 

Societal determinants in DR: 
- The high proportion of women and children infected suggests household transmission, perhaps through caregiving, shared items (linens, household surfaces), or close contact in domestic settings. 
- Stigma plays a role: The CDC authors of the DR study note that underreporting among MSM may happen because of homophobia and HIV stigma. 
- Healthcare access & diagnostics: limited PCR availability in some settings can hamper prompt diagnosis, increasing spread risk. 


Symptoms
- Prodromal symptoms: fever, intense headache, muscle aches/back pain, fatigue, swollen lymph nodes. 
- Rash: Nearly all (90.1%) confirmed cases in DR had a rash. 
- Lesion evolution: Lesions progress over time — macules → papules → vesicles → pustules → scabs/crusts. 
- Duration: Rash and lesions often last 2–4 weeks. 
- Complications: Potential for secondary bacterial infection of skin, dehydration, pain; more severe disease in immunocompromised individuals, children, or pregnant people. 
- Sequelae: Lesions can scar, and mucosal involvement (e.g., in mouth or genitals) may lead to more pain. 


Diagnosis
- Specimen collection: Samples are typically taken from lesion swabs (exudate, crust) and sometimes blood if no rash. 
- Laboratory testing: Polymerase chain reaction (PCR) is the gold standard to confirm mpox by detecting viral DNA. 
- Differential diagnosis: Because rash and fever are not specific, other infections must be ruled out (e.g., varicella (chickenpox), herpes, syphilis). 
- Clinical screening criteria: Using combinations of rash + fever + exposure history (risk profile) helps, especially in low-diagnostic-resource settings. In DR, relying only on rash and fever is limited due to low specificity. 
- Surveillance: In the DR, mpox is reportable, and health workers fill in standardized epidemiologic forms for suspected and confirmed cases. 
- Follow-up: Monitoring until all lesions scab over and heal, because contagiousness persists while lesions are active. WHO notes infectious period can last until scabs fall off. 



Biological Mechanisms (Pathogenesis & Physiology)
Virus & Structure:
- Mpox is caused by monkeypox virus (MPXV), an enveloped double-stranded DNA virus of the Orthopoxvirus genus (Poxviridae family). 
- Two major clades: Clade I (more historically associated with central Africa) and Clade II (with subclades IIa/IIb). The 2022–2023 global outbreak was driven by clade IIb. 

Transmission & Entry
- Transmission occurs via direct contact with lesions, bodily fluids (e.g., lesion exudate), respiratory droplets (close face-to-face), or contaminated objects (fomites like bedding). 
- In zoonotic settings, animal-to-human transmission may occur via bites, scratches, handling wild animals, or contact with their fluids. 
- The virus can enter through broken skin, mucous membranes (e.g., mouth, genitals), or respiratory tract (via close face-to-face contact).

Viral Replication & Spread
- After entering a host, MPXV replicates locally in epithelial cells (skin) and in lymphoid tissues, leading to viremia (virus in blood).
- It then seeds skin and mucosal surfaces, producing characteristic skin lesions (macules → papules → vesicles → pustules).

Immune Response
- The host’s innate immune system responds via type I interferons, cytokines, and the recruitment of immune cells to sites of infection.
- Adaptive immunity (T cells and B cells) mounts a response: cell-mediated immunity helps control infected cells; neutralizing antibodies help limit viral spread.
- Prior smallpox vaccination confers cross-protection because of immunological similarity (orthopoxviruses), which reduces susceptibility. 

Tissue/Organ Physiology Impact
- Skin: viral infection disrupts normal epidermal structure, leading to fluid-filled vesicles, pustules, and eventual scab formation.
- Lymph Nodes: viral replication in lymphoid tissues leads to lymphadenopathy (swollen nodes).
- Mucosal surfaces (oral, genital): lesions can impair mucosal integrity, potentially increasing risk of secondary infections or facilitating transmission.
- Systemic effects: viremia can lead to fever, malaise, myalgia, similar to systemic viral infections.


Persistence & Transmission Dynamics
- Lesions remain infectious until scabs fall off; the healing process can take weeks. 
- Because of the lesion progression and infectivity, both direct contact and indirect contact (via contaminated objects) are critical to transmission.
- In populations with limited immunity (e.g., no prior smallpox vaccination), more sustained or widespread transmission may occur.

In this analysis, the mpox infection data from the Democratic Republic of the Congo between 2023 and 2025 was examined using the Susceptible–Infected–Recovered (SIR) model. The goal was to estimate the infection rate (β) and recovery rate (γ) that best described the outbreak’s progression.
First, the dataset containing daily confirmed cases was loaded and used to calculate the cumulative number of cases over time. To approximate the number of currently infected individuals, a 14-day rolling window was applied under the assumption that most individuals recover within two weeks. These values were then normalized by an estimated total population of 1,000,000 to obtain fractions representing the susceptible, infected, and recovered portions of the population.
The SIR model was implemented using Euler’s method to numerically approximate the differential equations governing infection dynamics. A grid search was then performed across a range of β and γ values to minimize the sum of squared errors (SSE) between the model’s predicted I(t) and the observed I(t) from the data. The combination of parameters that produced the smallest SSE was chosen as the best fit.
After fitting the model, the predicted infection curve was compared to the actual data. The best-fit parameters were approximately β = 0.59 and γ = 0.48. The model captured the general trend of the epidemic, with a similar rise and fall in infection levels, but it overpredicted the height of the infection peak. This suggests that while the Euler’s method approximation effectively modeled the basic outbreak behavior, the assumption of constant β and γ does not fully capture real-world changes in transmission and recovery rates over time.
Overall, the SIR model provided a reasonable estimate of mpox transmission dynamics, and the SSE value served as a quantitative measure of the model’s prediction error.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


def convert_cumulative_to_SIR(cumulative_cases, N):
    """
    Convert cumulative case counts into S, I, R FRACTIONS of N.
    """
    cumulative_cases = np.array(cumulative_cases, dtype=float)

    # R(t): everyone ever infected, as a fraction of N
    R = cumulative_cases / N

    # I(t): new cases each time step, as a fraction of N
    new_cases = np.diff(cumulative_cases, prepend=cumulative_cases[0])
    I = new_cases / N

    # S(t): susceptible fraction
    S = 1 - I - R

    return S, I, R




csv_path = "/Users/zorawar/Desktop/CompBMEFolder/mpox_dem_rep_of_congo_data_2023_2025_new_cases.csv"
df = pd.read_csv(csv_path, parse_dates=["date"])

# cumulative confirmed cases
df["cum_cases"] = df["confirmed_cases"].cumsum()

# Choose N so that R(t) stays between 0 and 1 
N = df["cum_cases"].iloc[-1]

# Get S, I, R as fractions of N from the data
S_data, I_data, R_data = convert_cumulative_to_SIR(df["cum_cases"].values, N)
df["S"] = S_data
df["I"] = I_data
df["R"] = R_data


#  EULER SIR 

def euler_SIR(beta, gamma, S0, I0, R0, dt, n_steps):
    """
    Euler's method for SIR with S, I, R as fractions (N = 1):

        dS/dt = -beta * S * I
        dI/dt =  beta * S * I - gamma * I
        dR/dt =  gamma * I
    """
    S = np.zeros(n_steps)
    I = np.zeros(n_steps)
    R = np.zeros(n_steps)

    S[0], I[0], R[0] = S0, I0, R0

    for k in range(n_steps - 1):
        dS = -beta * S[k] * I[k]
        dI =  beta * S[k] * I[k] - gamma * I[k]
        dR =  gamma * I[k]

        S[k + 1] = S[k] + dt * dS
        I[k + 1] = I[k] + dt * dI
        R[k + 1] = R[k] + dt * dR

    return S, I, R




dt = 1.0
n_total = len(df)
mid_idx = n_total // 2          # start of prediction window
n_pred = n_total - mid_idx      # how many steps to predict

# Initial conditions at the MIDPOINT, from data
S0 = df.loc[mid_idx, "S"]
I0 = df.loc[mid_idx, "I"]
R0 = df.loc[mid_idx, "R"]

beta  = 0.27   # Guess and check Beta value
gamma = 0.2   



# Run Euler SIR starting at the midpoint, predicting forward
S_pred, I_pred, R_pred = euler_SIR(beta, gamma, S0, I0, R0, dt, n_pred)

# Build a full-length prediction array for plotting convenience
I_pred_full = np.full(n_total, np.nan)         # NaN for first half (no prediction there)
I_pred_full[mid_idx:] = I_pred                 # second half = model prediction



plt.figure(figsize=(14, 7))

# True I(t) from data (whole time window)
plt.plot(df["date"], df["I"], label="True I(t) from data", linewidth=2)

plt.plot(df["date"], I_pred_full, "--", label="SIR prediction (second half)", linewidth=2)


split_date = df["date"].iloc[mid_idx]
plt.axvline(split_date, color="k", linestyle=":", alpha=0.7)
plt.text(split_date, plt.ylim()[1]*0.95, "  prediction\n  starts here",
         va="top", ha="left")

plt.title(f"Predicting Second Half of I(t) Using SIR (beta={beta:.2f}, gamma={gamma:.2f})",
          fontsize=16)
plt.xlabel("Date", fontsize=14)
plt.ylabel("I(t) (fraction of population)", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 0. Load data and build "true" S, I, R

def convert_cumulative_to_SIR(cumulative_cases, N):
    """
    Convert cumulative case counts into S, I, R as FRACTIONS of N.
    """
    cumulative_cases = np.array(cumulative_cases, dtype=float)

    # R(t): everyone ever infected (fraction of N)
    R = cumulative_cases / N

    # I(t): new cases each step (fraction of N)
    new_cases = np.diff(cumulative_cases, prepend=cumulative_cases[0])
    I = new_cases / N

    # S(t): susceptible fraction
    S = 1 - I - R

    return S, I, R


csv_path = "/Users/zorawar/Desktop/CompBMEFolder/mpox_dem_rep_of_congo_data_2023_2025_new_cases.csv"
df = pd.read_csv(csv_path, parse_dates=["date"])

df["cum_cases"] = df["confirmed_cases"].cumsum()

# choose N as final cumulative count so S, I, R are in [0,1]
N = df["cum_cases"].iloc[-1]

S_data, I_data, R_data = convert_cumulative_to_SIR(df["cum_cases"].values, N)

df["S_data"] = S_data
df["I_data"] = I_data
df["R_data"] = R_data



# 1. Euler SIR 


def euler_SIR(beta, gamma, S0, I0, R0, dt, n_steps):
    """
    Euler's method for SIR with S, I, R as fractions (N = 1 implicitly):

        dS/dt = -beta * S * I
        dI/dt =  beta * S * I - gamma * I
        dR/dt =  gamma * I
    """
    S = np.zeros(n_steps)
    I = np.zeros(n_steps)
    R = np.zeros(n_steps)

    S[0], I[0], R[0] = S0, I0, R0

    for k in range(n_steps - 1):
        dS = -beta * S[k] * I[k]
        dI =  beta * S[k] * I[k] - gamma * I[k]
        dR =  gamma * I[k]

        S[k + 1] = S[k] + dt * dS
        I[k + 1] = I[k] + dt * dI
        R[k + 1] = R[k] + dt * dR

    return S, I, R


dt = 1.0
n_steps = len(df)


S0 = df["S_data"].iloc[0]
I0 = df["I_data"].iloc[0]
R0 = df["R_data"].iloc[0]



# 2. STEP 4 – SSE function

def sse_I(beta, gamma):
    """
    Sum of squared errors between model I(t) and data-based I(t),
    for a given beta, gamma.
    """
    S_model, I_model, R_model = euler_SIR(beta, gamma, S0, I0, R0, dt, n_steps)
    I_true = df["I_data"].values
    return np.sum((I_model - I_true) ** 2)


# quick check for one example pair 
example_sse = sse_I(0.5, 0.2)
print(f"Example SSE for beta=0.5, gamma=0.2: {example_sse:.6e}")


# 3. STEP 5 – grid search over beta, gamma

beta_grid   = np.linspace(0.05, 1.50, 40)   # 40 beta values
gamma_grid  = np.linspace(0.05, 1.00, 40)   # 40 gamma values

best_sse   = np.inf
best_beta  = None
best_gamma = None

for beta in beta_grid:
    for gamma in gamma_grid:
        current_sse = sse_I(beta, gamma)
        if current_sse < best_sse:
            best_sse = current_sse
            best_beta = beta
            best_gamma = gamma

print(f"\nBest-fit parameters from grid search:")
print(f"  beta  = {best_beta:.4f}")
print(f"  gamma = {best_gamma:.4f}")
print(f"  SSE   = {best_sse:.6e}")


# 4. Plot best-fit model vs. data I(t)

S_best, I_best, R_best = euler_SIR(best_beta, best_gamma, S0, I0, R0, dt, n_steps)
I_true = df["I_data"].values

plt.figure(figsize=(14, 7))
plt.plot(df["date"], I_true, label="True I(t) from data", linewidth=2)
plt.plot(df["date"], I_best, "--", label="Best-fit SIR model I(t)", linewidth=2)

plt.title(
    f"I(t): Data vs Best-Fit SIR Model\nbeta={best_beta:.3f}, gamma={best_gamma:.3f}",
    fontsize=16
)
plt.xlabel("Date", fontsize=14)
plt.ylabel("I(t) (fraction of population)", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Load data
csv_path = "/Users/zorawar/Desktop/CompBMEFolder/mpox_dem_rep_of_congo_data_2023_2025_new_cases.csv"
df = pd.read_csv(csv_path, parse_dates=["date"])
df["cum_cases"] = df["confirmed_cases"].cumsum()

#  Population estimate 
N = 1_000_000  # adjust if known

#  Estimate currently infected (active) using rolling window 
infection_period = 14  # assume 14 days average infection duration
df["active_cases"] = df["confirmed_cases"].rolling(window=infection_period, min_periods=1).sum()

# Fractions of population
df["I_data"] = df["active_cases"] / N
df["R_data"] = df["cum_cases"] / N
df["S_data"] = 1 - df["I_data"] - df["R_data"]

# Quick plots of S, I, R 
plt.figure(figsize=(14, 4))
plt.plot(df["date"], df["S_data"])
plt.title("Susceptible (S) Approximation")
plt.xlabel("Date")
plt.ylabel("Fraction")
plt.grid(True, linestyle="--", alpha=0.6)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 4))
plt.plot(df["date"], df["I_data"])
plt.title("Infectious (I) Approximation")
plt.xlabel("Date")
plt.ylabel("Fraction")
plt.grid(True, linestyle="--", alpha=0.6)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 4))
plt.plot(df["date"], df["R_data"])
plt.title("Recovered (R) Approximation")
plt.xlabel("Date")
plt.ylabel("Fraction")
plt.grid(True, linestyle="--", alpha=0.6)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# SIR model (Euler)
def euler_SIR(beta, gamma, S0, I0, R0, dt, n_steps):
    S = np.zeros(n_steps)
    I = np.zeros(n_steps)
    R = np.zeros(n_steps)
    S[0], I[0], R[0] = S0, I0, R0

    for k in range(n_steps - 1):
        dS = -beta * S[k] * I[k]
        dI = beta * S[k] * I[k] - gamma * I[k]
        dR = gamma * I[k]
        S[k + 1] = S[k] + dt * dS
        I[k + 1] = I[k] + dt * dI
        R[k + 1] = R[k] + dt * dR
    return S, I, R


# Common time/grid stuff
dt = 1.0
n_total = len(df)
mid_idx = n_total // 2

# Split data for Step 2 & 3
df_train = df.iloc[:mid_idx]
df_test = df.iloc[mid_idx:]
n_train = len(df_train)
n_test = len(df_test)

# Initial conditions from first data point
S0 = df_train["S_data"].iloc[0]
I0 = df_train["I_data"].iloc[0]
R0 = df_train["R_data"].iloc[0]

I_true_full = df["I_data"].values
I_true_train = df_train["I_data"].values
I_true_test = df_test["I_data"].values

# Parameter search grid
beta_vals = np.linspace(0.05, 1.0, 40)
gamma_vals = np.linspace(0.05, 0.5, 40)


# STEP 1: Fit beta,gamma on FULL data (Euler)
def sse_I_full(beta, gamma):
    _, I_model, _ = euler_SIR(beta, gamma, S0, I0, R0, dt, n_total)
    return np.sum((I_model - I_true_full) ** 2)

best_sse_full = np.inf
best_beta_full = None
best_gamma_full = None

for beta in beta_vals:
    for gamma in gamma_vals:
        sse = sse_I_full(beta, gamma)
        if sse < best_sse_full:
            best_sse_full = sse
            best_beta_full = beta
            best_gamma_full = gamma

print(f"STEP 1 (full data) – best β={best_beta_full:.3f}, γ={best_gamma_full:.3f}")
print(f"STEP 1 training SSE (full data): {best_sse_full:.3e}")

S_full_pred, I_full_pred, R_full_pred = euler_SIR(
    best_beta_full, best_gamma_full, S0, I0, R0, dt, n_total
)

plt.figure(figsize=(14, 7))
plt.plot(df["date"], I_true_full, label="True I_data", linewidth=2)
plt.plot(df["date"], I_full_pred, "--", color="orange",
         label=f"Euler SIR (full fit β={best_beta_full:.3f}, γ={best_gamma_full:.3f})", linewidth=2)
plt.title("STEP 1: Euler SIR – Fit on Full Dataset")
plt.xlabel("Date")
plt.ylabel("I(t) (fraction of population)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()



# STEP 2: Your code – train on first half, test on second half (Euler)
def sse_I_train(beta, gamma):
    S, I, R = euler_SIR(beta, gamma, S0, I0, R0, dt, n_train)
    return np.sum((I - I_true_train) ** 2)

best_sse_half = np.inf
best_beta_half = None
best_gamma_half = None

for beta in beta_vals:
    for gamma in gamma_vals:
        sse = sse_I_train(beta, gamma)
        if sse < best_sse_half:
            best_sse_half, best_beta_half, best_gamma_half = sse, beta, gamma

print(f"\nSTEP 2 (first half) – best β={best_beta_half:.3f}, γ={best_gamma_half:.3f}")
print(f"STEP 2 training SSE (first half): {best_sse_half:.3e}")

# Predict full series using best-fit parameters from first half
S_pred, I_pred, R_pred = euler_SIR(best_beta_half, best_gamma_half, S0, I0, R0, dt, n_total)

# Compute SSE on test (second half)
I_pred_test = I_pred[mid_idx:]
sse_test_half = np.sum((I_pred_test - I_true_test) ** 2)
print(f"STEP 2 test SSE (second half) = {sse_test_half:.3e}")

# Save Euler first-half-fit prediction for comparison in Step 3
I_euler_half = I_pred.copy()

# Plot Step 2
plt.figure(figsize=(14, 7))
plt.plot(df["date"], df["I_data"], label="True I(t) data", linewidth=2)
plt.plot(df["date"], I_pred, "--", color="orange",
         label=f"SIR model (Euler, β={best_beta_half:.3f}, γ={best_gamma_half:.3f})", linewidth=2)

split_date = df["date"].iloc[mid_idx]
plt.axvline(split_date, color="k", linestyle=":", alpha=0.7)
plt.text(split_date, plt.ylim()[1]*0.95, "  Prediction starts here",
         va="top", ha="left", fontsize=11)

plt.title("STEP 2: Euler SIR – Trained on First Half, Tested on Second Half", fontsize=16)
plt.xlabel("Date", fontsize=14)
plt.ylabel("I(t) (fraction of population)", fontsize=14)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# STEP 3: RK (solve_ivp), fit on first half, compare Euler vs RK

def sir_rhs(t, y, beta, gamma):
    S, I, R = y
    dSdt = -beta * S * I
    dIdt = beta * S * I - gamma * I
    dRdt = gamma * I
    return [dSdt, dIdt, dRdt]


def rk_SIR(beta, gamma, S0, I0, R0, dt, n_steps):
    t_span = (0.0, (n_steps - 1) * dt)
    t_eval = np.arange(0.0, n_steps * dt, dt)

    sol = solve_ivp(
        fun=lambda t, y: sir_rhs(t, y, beta, gamma),
        t_span=t_span,
        y0=[S0, I0, R0],
        t_eval=t_eval,
        method="RK45"
    )

    S, I, R = sol.y
    return S, I, R


def sse_I_train_rk(beta, gamma):
    S_rk, I_rk, R_rk = rk_SIR(beta, gamma, S0, I0, R0, dt, n_total)
    return np.sum((I_rk[:n_train] - I_true_train) ** 2)

best_sse_rk = np.inf
best_beta_rk = None
best_gamma_rk = None

for beta in beta_vals:
    for gamma in gamma_vals:
        sse = sse_I_train_rk(beta, gamma)
        if sse < best_sse_rk:
            best_sse_rk, best_beta_rk, best_gamma_rk = sse, beta, gamma

print(f"\nSTEP 3 (RK first half) – best β={best_beta_rk:.3f}, γ={best_gamma_rk:.3f}")
print(f"STEP 3 training SSE (RK, first half): {best_sse_rk:.3e}")

# Predict full series with RK
S_rk_pred, I_rk_pred, R_rk_pred = rk_SIR(best_beta_rk, best_gamma_rk, S0, I0, R0, dt, n_total)

# SSE on test (second half) for RK
I_rk_test = I_rk_pred[mid_idx:]
sse_test_rk = np.sum((I_rk_test - I_true_test) ** 2)
print(f"STEP 3 test SSE (RK, second half) = {sse_test_rk:.3e}")

# Plot comparison: data vs Euler-vs-RK (both fit on first half)
plt.figure(figsize=(14, 7))
plt.plot(df["date"], I_true_full, "k-", linewidth=2, label="True I_data")
plt.plot(df["date"], I_euler_half, "-", linewidth=2, label="Euler (fit first half)")
plt.plot(df["date"], I_rk_pred, "--", linewidth=2, label="RK (fit first half)")

plt.axvline(split_date, color="k", linestyle=":", alpha=0.7, label="Train/Test split")
plt.title("STEP 3: Euler vs RK – Both Fit on First Half, Predict Second Half")
plt.xlabel("Date")
plt.ylabel("I(t) (fraction of population)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- Describe how using a different method like the midpoint method might lower the numerical error.

In our implementation we use Euler’s method, which is a first-order method. The global error over the whole simulation is on the order of O(h)O(h), where hh is the step size (in our case, 1 day). Because Euler only uses the slope at the beginning of each time step, it can overshoot or undershoot the true solution, especially when the infection curve is changing rapidly.

A midpoint method  effectively takes an extra “preview” step: it first estimates the slope at the midpoint of the interval and then uses that midpoint slope to update the solution. This makes the method second-order accurate, with global error on the order of 
O(h2)O(h 2 ). For the same step size, the midpoint method will therefore produce a trajectory for S(t), I(t), and R(t) that stays closer to the exact solution of the SIR equations than Euler’s method.
In practice, this means that for the same daily time step, a midpoint method would:
Reduce numerical drift in S, I, and R over long time intervals
Better capture the shape and timing of the infection peak
Lower the discrepancy between the simulated I(t) and the data without changing β or γ
So even if we keep the same parameters (β, γ), switching from Euler to a midpoint method would typically lower the numerical error in the SIR simulation and could reduce the SSE between the model and the data.
